```python
import numpy as np
import matplotlib.pyplot as plt
import sys 
import matplotlib as mpl
%matplotlib inline

mpl.rcParams['xtick.labelsize'] = 15
mpl.rcParams['ytick.labelsize'] = 15
mpl.rcParams['axes.labelsize'] = 20
mpl.rcParams['legend.fontsize'] = 15
mpl.rcParams['figure.figsize']  = (9, 6)
```

## Central line

```python
import sys
sys.path.append('./Lyn_exp_1995/')
sys.path.append('./Sohankar_LES_1999/')
```

```python
umean_lyn = np.loadtxt('./Lyn_exp_1995/CentralLine_exp_umean.csv', delimiter=',')
x_lyn = umean_lyn[:,0]
y_lyn = umean_lyn[:,1]

umean_ssm = np.loadtxt('./Sohankar_LES_1999/CentralLine_SSM_Smagorinsky_umean.csv', delimiter=',')
x_ssm = umean_ssm[:,0]
y_ssm = umean_ssm[:,1]

umean_dsm = np.loadtxt('./Sohankar_LES_1999/CentralLine_DSM_dynamic_Smag_umean.csv', delimiter=',')
x_dsm = umean_dsm[:,0]
y_dsm = umean_dsm[:,1]


plt.figure()
plt.plot(x_lyn, y_lyn, 'D', label='Lyn-exp (1995)')
plt.plot(x_ssm, y_ssm, label='Sohankar-SSM-LES (2000)')
plt.plot(x_dsm, y_dsm, label='Sohankar-DSM-LES (2000)')

plt.legend()
plt.xlabel('$x/d$')
plt.ylabel('$<u>/U_0$')
```

```python
uRMS_lyn = np.loadtxt('./Lyn_exp_1995/CentralLine_exp_upRMS.csv', delimiter=',')
x_lyn = uRMS_lyn[:,0]
y_lyn = uRMS_lyn[:,1]

uRMS_ssm = np.loadtxt('./Sohankar_LES_1999/CentralLine_SSM_upRMS.csv', delimiter=',')
x_ssm = uRMS_ssm[:,0]
y_ssm = uRMS_ssm[:,1]

uRMS_dsm = np.loadtxt('./Sohankar_LES_1999/CentralLine_DSM_upRMS.csv', delimiter=',')
x_dsm = uRMS_dsm[:,0]
y_dsm = uRMS_dsm[:,1]


plt.figure()
plt.plot(x_lyn, y_lyn, 'D', label='Lyn-exp (1995)')
plt.plot(x_ssm, y_ssm, label='Sohankar-SSM-LES (2000)')
plt.plot(x_dsm, y_dsm, label='Sohankar-DSM-LES (2000)')

plt.legend()
plt.xlabel('$x/d$')
plt.ylabel("$\sqrt{<u'u'>/U_0^2}$")
```

```python
vRMS_lyn = np.loadtxt('./Lyn_exp_1995/CentralLine_exp_vpRMS.csv', delimiter=',')
x_lyn = vRMS_lyn[:,0]
y_lyn = vRMS_lyn[:,1]

vRMS_ssm = np.loadtxt('./Sohankar_LES_1999/CentralLine_SSM_vpRMS.csv', delimiter=',')
x_ssm = vRMS_ssm[:,0]
y_ssm = vRMS_ssm[:,1]

vRMS_dsm = np.loadtxt('./Sohankar_LES_1999/CentralLine_DSM_vpRMS.csv', delimiter=',')
x_dsm = vRMS_dsm[:,0]
y_dsm = vRMS_dsm[:,1]


plt.figure()
plt.plot(x_lyn, y_lyn, 'D', label='Lyn-exp (1995)')
plt.plot(x_ssm, y_ssm, label='Sohankar-SSM-LES (2000)')
plt.plot(x_dsm, y_dsm, label='Sohankar-DSM-LES (2000)')

plt.legend()
plt.xlabel('$x/d$')
plt.ylabel("$\sqrt{<v'v'>/U_0^2}$")
```

```python
wRMS_ssm = np.loadtxt('./Sohankar_LES_1999/CentralLine_SSM_wpRMS.csv', delimiter=',')
x_ssm = wRMS_ssm[:,0]
y_ssm = wRMS_ssm[:,1]

wRMS_dsm = np.loadtxt('./Sohankar_LES_1999/CentralLine_DSM_wpRMS.csv', delimiter=',')
x_dsm = wRMS_dsm[:,0]
y_dsm = wRMS_dsm[:,1]


plt.figure()
plt.plot(x_ssm, y_ssm, '^', label='Sohankar-SSM-LES (2000)')
plt.plot(x_dsm, y_dsm, 'o', label='Sohankar-DSM-LES (2000)')

plt.legend()
plt.xlabel('$x/d$')
plt.ylabel("$\sqrt{<w'w'>/U_0^2}$")
```
